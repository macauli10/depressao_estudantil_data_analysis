Convertendo a coluna de duracao de sono para float para uma melhor análise sem comprometer a integridade dos dados

```python

def converter_sono(tempo):
    if pd.isnull(tempo):
        return None
    tempo = tempo.strip().lower()
    if "less than 5" in tempo:
        return 4.5
    elif "5-6" in tempo:
        return 5.5
    elif "7-8" in tempo:
        return 7.5
    elif "more than 8" in tempo:
        return 9
    else:
        return None

df['Sono_horas'] = df['Duração_do_Sono'].apply(converter_sono)
```
